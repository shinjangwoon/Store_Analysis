```python
import numpy as np

def get_cohesion(word):
    # 글자가 아니거가 공백, 혹은 희귀한 단어인 경우
    if(not word) or((word in L) == False):
        return 0.0
    
    if len(word) == 1:
        return 1.0
    
    word_freq = L.get(word, 0)
    base_freq = L.get(word[:1], 0)
    
    if base_freq == 0:
        return 0.0
    else:
        return np.power((word_freq / base_freq), 1/(len(word) - 1))
    
print('청와대 : ', get_cohesion('청와대'))
```
