### 시간 칼럼 처리

```python
def time_column_preprocess(df):
    """시간 관련 칼럼 정리"""
    ttm_len = df["REAL_SAL_TTM"].astype(int).astype(str).apply(len)
    display(ttm_len.value_counts().sort_index())
    
    print(f"REAL_SAL_TTM 4글자 이하 제거: {(ttm_len <= 4).sum():,}")
    df = df[ttm_len >= 5]

    dt_col = df["REAL_SAL_DT"].astype(str) + df["REAL_SAL_TTM"].astype(int).astype(str).str.zfill(6)
    assert (dt_col.apply(len) != 14).sum() == 0
    df = df.assign(log_dt = pd.to_datetime(dt_col, format="%Y%m%d%H%M%S"))

    df = df.drop(["REAL_SAL_DT", "REAL_SAL_TTM", "SAL_DT"], axis=1)
    
    return df
```

```python
%%time
df = time_column_preprocess(df)
```

```python
%%time
## 칼럼 정리 추가
df = df.drop(["POS_NO", "TRAN_NO"], axis=1)
df = df.rename(columns={"마트명":"mart_name"})
```

### 카테고리 매핑 정답지

```python
cate_map = pd.read_csv(path+"top3000_3.csv")
cate_raw = pd.read_excel(path+"리테일_카테고리_분류.xlsx")
```

```python
def cate_merge(cate_map, cate_raw):
    
    cate_map = cate_map[["GDS_NM", "대분류", "소분류"]]
    cate_map = cate_map.rename(columns={"대분류":"대", "소분류":"소"})

    cate_raw = cate_raw[["대", "소"]].drop_duplicates()
    cate_raw = cate_raw.assign(cate_id = range(len(cate_raw)))
    
    cate_df = cate_raw.merge(cate_map, on=["대", "소"], how="left")

    return cate_df
```

```python
cate_df = cate_merge(cate_map, cate_raw)
```

```python
len(cate_df.cate_id.unique())
```

```python
cate_df.to_parquet(path+"category_map2.parquet")
```

```python
%%time
## 매핑
cate_map_dict = cate_df.dropna().set_index("GDS_NM").cate_id.to_dict()
df = df.assign(cate_id = df.GDS_NM.map(cate_map_dict))
```
